```python
import streamlit as st
import pandas as pd
import altair as alt

# Load data
df = pd.read_csv('./Modded_Datasheet.csv')

# Create scatter plot
scatter_plot = alt.Chart(df).mark_circle().encode(
    x='Pushups_Sets',
    y='Pushups_Repetitions',
    tooltip=['Pushups_Sets', 'Pushups_Repetitions'] # corrected typo here
).interactive()

# Display scatter plot
st.write(scatter_plot)

plt.show(scatter_plot)
```
